```python
import json
import xml.etree.ElementTree as xml
import cv2

t=open("/home/rahul/temp_data/train/images/train.json")
rr=json.load(t)
for i in rr:
   # print(rr[i]['filename'])
   
   
    root=xml.Element("annotation")
    folder=xml.Element("folder")
    folder.text="images"
    root.append(folder)
   
    filename=xml.Element("filename")
    filename.text=rr[i]['filename']
    root.append(filename)
   
    path=xml.Element('path')
    path.text="/home/rahul/temp_data/train/images/"+rr[i]['filename']
    root.append(path)
    print(path.text)
    source=xml.Element("source")
    database=xml.SubElement(source,"database")
    database.text="Unknown"
    root.append(source)
    size=xml.Element("size")
    
    #print(ss
    ss="/home/rahul/temp_data/train/images/"+rr[i]['filename']
    print(ss)
    img=cv2.imread(ss)
    height1, width1, channels = img.shape
    #print(height)
    width=xml.SubElement(size,"width")
    width.text=str(width1)
    height=xml.SubElement(size,"height")
    height.text=str(height1)
    depth=xml.SubElement(size,"depth")
    depth.text=str(channels)
    root.append(size)
    segmented=xml.Element("segmented")
    segmented.text=str(0)
    root.append(segmented)
    tt=rr[i]['regions']
    #print(rr[i]['filename'])
    for j in tt:
               
        lx=j['shape_attributes']['all_points_x']
        ly=j['shape_attributes']['all_points_y']
       
        x_min=min(lx)
        x_max=max(lx)
        y_min=min(ly)
        y_max=max(ly)
   # print(i)
        object=xml.Element("object")
   
        name=xml.SubElement(object,"name")
        name.text=j['region_attributes']['onion']
        pose= xml.SubElement(object,"pose")
        pose.text="Unspecified"
        truncated=xml.SubElement(object,"truncated")
        truncated.text=str(0)
        difficult=xml.SubElement(object,"difficult")
        difficult.text=str(0)
        bndbox= xml.SubElement(object,"bndbox")
        xmin=xml.SubElement(bndbox,"xmin")
        xmin.text=str(x_min)
        ymin=xml.SubElement(bndbox,"ymin")
        ymin.text=str(y_min)
        xmax=xml.SubElement(bndbox,"xmax")
        xmax.text=str(x_max)
        ymax=xml.SubElement(bndbox,"ymax")
        ymax.text=str(y_max)
       
       
       

        #print(lx)
        #print(ly)
        #print(x_min)
        #print(x_max)
        #print(y_min)
        #print(y_max)
   

        #print(j['region_attributes']['onion'])
        #print()
        root.append(object)
   
   
   
   

   
   
   
   
   
    tree=xml.ElementTree(root)
    name=rr[i]['filename']
    jj=name.split('.')[0]
    #print(jj)
    path="/home/rahul/temp_data/train/annotations/"
    path+=jj+'.xml'
    print(path)
   
    with open(path,"wb") as files:
        tree.write(files)

   
   
    
```
